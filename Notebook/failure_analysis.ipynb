```python
import pandas as pd

df = pd.read_csv(
    r"C:\Users\ankit\Predictive-Maintenance-IoT-AI\Data\preprocessed_ai4i2020.csv"
)

print(df.shape)
df.head()
```

```python
# Machine failure count
print(df["Machine failure"].value_counts())
```

```python
# Percentage
print(
    df["Machine failure"]
    .value_counts(normalize=True) * 100
)
```

```python
# Visualization
import matplotlib.pyplot as plt

df["Machine failure"].value_counts().plot(kind="bar")

plt.title("Machine Failure Distribution")
plt.xlabel("Machine Failure")
plt.ylabel("Count")

plt.show()
```
